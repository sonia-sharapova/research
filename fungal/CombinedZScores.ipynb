```python
import csv

fullScoreList = []
newRow = []
count = 0


    
with open('/Users/Sonia/mergedUniqueGenes.csv', 'r') as fullGenes:
    allGenes = fullGenes.readlines()
    
    with open('/Users/Sonia/zScoresSPathsUpper.csv', 'r') as Pathways:
        path = Pathways.readlines()
    
        for gene in allGenes:
            gene.rstrip('\r\n')
            el = gene.split(',')
            for y in el:
                newRow.append(y.rstrip('\r\n').strip('\"'))
            
            
            if newRow != ['']:
                #print(newRow)
                for pth in path:
                    pthEl = pth.split(',')
                
                    #print(newRow[0])
                    #print(pthEl[0])
                
                    if pthEl[0] == newRow[0]:
                        print(gene)
                        newRow.append(pthEl[1])
                        count = 1
                if count == 0:
                    newRow.append(0)
                fullScoreList.append(newRow)
        
            count = 0
            newRow = []             
            #print(newRow)
        
#print(fullScoreList)

with open('CombinedZScores.csv', 'w') as outFile:
    
    write = csv.writer(outFile)
      
    #write1.writerow(fields)
    write.writerows(fullScoreList)
    
    
    #for x in fullScoreList:
        #x = str(x)
        #print(x)
        #outFile.write(x)
                    
```
