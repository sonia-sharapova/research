```python
def ranks(rank):
    #rank[:,-1]
    for row in rank:
        print(row[-1])
        if row[-1] == "BRCA_Basal":
            row[-1] = 1
        elif row[-1] == "BRCA_Her2":
            row[-1] = 2
        elif row[-1] == "BRCA_LumA":
            row[-1] = 3
        elif row[-1] == "BRCA_LumB":
            row[-1] = 4
        else:
            row[-1] = 0
        print(row[-1])
    return rank
  
        
        
```

```python
import math

def remNan(data):

    np_data = data


    for row in np_data[1:,1:]:
        ind = 0
        for i in row:
            if math.isnan(float(i)):
                row[ind] = 0.0
            elif float(i) < 0:
                row[ind] = 0.0
            ind = ind + 1
    return np_data
```

```python
prot = remNan(prot)
prot[0][-1] = "Cancer Type"
pd_data_prot = pd.DataFrame(prot)
```

```python
def remSplit(data,dataset):
    

    newRow2 = []
    sep = '|'


    for x in dataset:
        if type(x) == str:
            stripped = x.split(sep, 1)[0]
        else:
            stripped = x
        newRow2.append(stripped)
    print(newRow2)
    data.loc[0] = newRow2
    return data
```
