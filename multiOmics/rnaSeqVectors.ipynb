```python
def load_dataset(filename, features):
    data = pd.read_csv(filename, sep=',', header=None)
    data = handle_non_numerical_data(data)
    points = data.values
    labels = points[0,0]
    dataset = remNan(points)
    
    dataset = np.array(dataset)
    
    X = dataset[1:, 1:-1]
    y = dataset[1:, -1]
    return X, y

    

def remNan(filename):
    filename = allInt(filename)
    #newDataset = removeHeader(filename)
    newData = []
    for row in filename:
        for x in range(len(row)):
            if (str(row[x]) == "NA" or str(row[x]) == "nan" or str(row[x]) == "NaN"):
                row[x] = '0'
        newData.append(row)
    newData = pd.DataFrame(newData, columns =[newData[0]])
    return newData

def allInt(filename):
    newData = []
    for row in filename:
        for x in range(len(row)):
            row[x] = str(row[x])
        newData.append(row)
    return newData

"""def removeHeader(data):
    print(data[:][1])
    newData = data[:][1:]
    return newData"""
        
def compute_accuracy(y_true, y_pred):
    correct_predictions = 0
    for true, predicted in zip(y_true, y_pred):
        if true == predicted:
            correct_predictions += 1
    accuracy = correct_predictions/len(y_true)
    return accuracy


def getFeatureLength(csv):
    with open(csv, 'r') as read_obj:
        csv_reader = reader(read_obj)
        readFile = list(csv_reader)
        featureLength = len(readFile[0])
    return featureLength

def createList(n):
    lst = []
    for i in range(n+1):
        lst.append(i)
    return(lst)

def getTrainTest(csv, features):
    X, y = load_dataset(csv, features)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=False)
    features = y
    
    return X_train, X_test, y_train, y_test

def handle_non_numerical_data(df):
    column = df.iloc[:,-1:]
    
    text_digit_vals = {}
    def convert_to_int(val):
        return text_digit_vals[val]
    
    column_contents = []
    cc = df.iloc[:,-1:].values.tolist()
    for a in cc:
        column_contents.append(tuple(a))
        
    unique_elements = set(column_contents)
    x = 0
    for unique in unique_elements:
        if unique not in text_digit_vals:
            text_digit_vals[unique] = x
            x+=1
            
    asNum = [text_digit_vals[tuple(b)] for b in cc]
    
    #df.drop(df.iloc[:,-1:], axis = 1, inplace = True)
    a_series = pd.Series(asNum)
    a_series = a_series.to_frame()

    df.iloc[:,-1:] = a_series
    return df
```
