```python
import numpy as np
import pandas as pd
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

 
# load the dataset
def load_dataset(filename, features):
    # load the dataset as a pandas DataFrame
    data = read_csv(filename, header=None)
    dataset = data.values
    #print(dataset.shape)
    #print(dataset[:,features])
    labels = dataset[0][features]
    
    

    newDataset = dataset[:,features]
    # split into input (X) and output (y) variables
    
    #print(newDataset)

    newDataset = removeHeader(newDataset)
    
    #print(newDataset)

    X = newDataset[:, :-1]
    y = newDataset[:,-1]
    

    return X, y



def getTrainTest(csv, features):
    X, y = load_dataset(csv, features)
# split into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=False)
    features = y
    
    return X_train, X_test, y_train, y_test




from sklearn.linear_model import LinearRegression

#getTrainTest('/Users/soniasharapova/Documents/FatiResearch/dummyDataset.csv', features)

def predictions(csv, feats):
    X_train, X_test, y_train, y_test = getTrainTest(csv, feats)

    #clf = LinearRegression()
    model = RandomForestRegressor()
    model.fit(X_train, y_train)

    #clf.fit(X_train, y_train)
    #print("LinearRegression")
    y_pred = model.predict(X_test)


    #print(y_pred)
    #print(y_test)

    return(np.linalg.norm(y_pred - np.array(y_test).astype(float)))


def getFeats(filename):
    # load the dataset as a pandas DataFrame
    data = read_csv(filename, header=None)
    dataset = data.values
    
    features = dataset[0]
    #print(features)
    return features

def convertLabels(csv, features):
    data = read_csv(csv, header=None)
    dataset = data.values
    #print(dataset.shape)
    #print(dataset[:,features])
    labels = dataset[0][features]
    
    #newDataset = dataset[:,features]
    
    print(features)
    print(labels)
    
def removeHeader(data):
    newData = data[1:]
    return newData
 
import matplotlib.pyplot as plt 
import pandas as pd 


def bestFeats(csv, feats):
    featurePred=[]
    featsCopy = feats
    #print(feats)
    #print(featsCopy)
    length = len(feats)
    for x in range(length):
        
        if(x == 0):
            featsCopy = featsCopy[(x+1):]
            #print(featsCopy)

            featurePred.append(predictions(csv, featsCopy))
            featsCopy = []
            featsCopy = feats
        else:
            featsCopy = featsCopy[:x] + featsCopy[(x+1):]
            #print(featsCopy)

            featurePred.append(predictions(csv, featsCopy))
            featsCopy = []
            featsCopy = feats
        
        #sort 

        #print(featurePred)
        #print(feats)

    #featurePred shows the prediction if this feature was removed. If score is low, this means prediction would be lower if this was removed
    
    L = [ (featurePred[i],i) for i in range(len(featurePred)) ]

    L.sort()
    #print(L)
    sorted_l,permutation = zip(*L)
    #print(sorted_l)
    #print(permutation)
    permutationNp = np.array(permutation)
    #print(featurePred)
    #print(permutationNp)
    
    #print(predictions(csv, feats))
    
    for y in range(length - 1):
        featureSubset = permutationNp[y:]
        predIter = predictions(csv, featureSubset)

    #print(featurePred)
    #print(permutationNp)

    return permutationNp
    

def createList(n):
    lst = []
    for i in range(n+1):
        lst.append(i)
    return(lst)

#features = createList(384)
mirnaFeatures = createList(184)
mrnaFeatures = createList(200)

#features = getFeats('/Users/soniasharapova/Documents/TrainTestComb.csv')
#385 total

preds = bestFeats('/Users/soniasharapova/Documents/mirnaCategorical.csv', mirnaFeatures)
convertLabels('/Users/soniasharapova/Documents/mirnaCategorical.csv', preds)
```
